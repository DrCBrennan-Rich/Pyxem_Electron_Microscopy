```python
#Switch matplotlib to display figures in pop-out mode
%matplotlib qt 

import hyperspy.api as hs
import gc
```

```python
#Load the data and plot
RawData = hs.load(r"C:\Users\YourPathHere", reader="hspy")
RawData.plot()
```

```python
#Assign type as electron diffraction so pyxem can work
RawData.set_signal_type('electron_diffraction')  

#Rebin if required
#RawData = RawData.rebin(scale=(1, 1, 2, 2))
```

```python
#Inspect the size and units of the data
RawData.axes_manager
```

```python
#Assign data to 32-bit floats to save a bit of memory 
RawData.data = RawData.data.astype('float32')

#Normalise the data
RawData.data *= 1 / RawData.data.max()

#Removes variables that no longer have proper assigment to free up some memory
gc.collect()
```
