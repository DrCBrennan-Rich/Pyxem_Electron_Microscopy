```python
%matplotlib qt

import numpy as np
import diffpy.structure
import matplotlib.pyplot as plt
from diffsims.generators.diffraction_generator import DiffractionGenerator
from diffsims.generators.zap_map_generator import get_rotation_from_z_to_direction

plt.rcParams.update({'font.size': 14})
```

```python
#Import the structures in the form of cif files, 

structure_Al2O3 = diffpy.structure.loadStructure("Al2O3.cif")
structure_Fe3S2 = diffpy.structure.loadStructure("Fe3Sn2.cif")
structure_Fe3S2 = diffpy.structure.loadStructure("Mn3Ge_Tetragonal.cif")
```

```python
#Select the structure we want to simulate

Structure = structure_Al2O3
```

```python
#This produces the euler angles for the rotations of the crystal from the incident beam direction [uvw]
u, v, w = 0,0,1
selected_rot = get_rotation_from_z_to_direction(Structure,(u, v, w))
get_rotation_from_z_to_direction(Structure,(u, v, w))
```

```python
ReciprocalRadius = 1. #Radius to simulate out to in A^-1
phi1 = 0  # Define phi1, the rotation in plane, here
diff_gen = DiffractionGenerator(accelerating_voltage = 100.,
                                precession_angle = 0.,
                                scattering_params = 'xtables',
                                shape_factor_model = "linear",
                                minimum_intensity = 0.000001,
                                )
Simulate = diff_gen.calculate_ed_data(Structure, 
                                reciprocal_radius = ReciprocalRadius,
                                rotation = (phi1, selected_rot[1],selected_rot[2]), 
                                with_direct_beam = True,
                                max_excitation_error = 0.03 #Not an error, actually the "smearing" of the repricol lattice points
                                )                           #so the Ewald sphere will actually intersect the points
                                                           
```

```python
fig, axs = plt.subplots()
axs.set_aspect("equal")
Simulate.plot(size_factor=5, 
         show_labels=False, #Turns on or off the (hkl) values
         ax=axs, 
         color ='red'
        )
#Choose if a coordinate system is drawn round the spots
plt.axis('on')
plt.rcParams['axes.facecolor'] = 'white'
axs.text(0.85, 0.05, f"[{u} {v} {w}]", fontsize=20, ha='center', transform=axs.transAxes)
plt.xlabel('kx Axis ($\mathregular{A^{-1}}$)', fontsize=20)
plt.ylabel('ky Axis ($\mathregular{A^{-1}}$)', fontsize=20)
```
